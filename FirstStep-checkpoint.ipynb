## Here is the first program regarding $Sin(x)$ function. 


This program will plot some graphs by taking some functions.

$$f(x)=\frac{\sin(x)}{x}$$

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import numpy as np
import matplotlib.pyplot as plt
pi = np.pi
x = np.linspace(-4*pi, 4*pi, 1000)
plt.rcParams['figure.figsize']=(12,6)
plt.rcParams.update({'font.size': 17})
plt.axhline(0, color='green', lw=2)
plt.axvline(0, color='yellow', lw=2)# This will add a verticle line, from zero location, color yellow and line width 
plt.xticks([-4*pi,-2*pi,0,2*pi,4*pi], ['$-4\pi$','$-2\pi$','$0$','$2\pi$','$4\pi$'])
plt.plot(x,np.sin(x)/x, label=r'$f(x)=\frac{\sin(x)}{x}$')
plt.legend(loc='best', fontsize=20)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(12,6)
plt.rcParams.update({'font.size': 17})
plt.axhline(0, color='black', lw=2)
plt.xticks([-4*pi,-2*pi,0,2*pi,4*pi], ['$-4\pi$','$-2\pi$','$0$','$2\pi$','$4\pi$'])
plt.plot(x,np.cos(x)/x, label=r'$f(x)=\frac{\cos(x)}{x}$')
plt.legend(loc='best', fontsize=20)
plt.show()
```
